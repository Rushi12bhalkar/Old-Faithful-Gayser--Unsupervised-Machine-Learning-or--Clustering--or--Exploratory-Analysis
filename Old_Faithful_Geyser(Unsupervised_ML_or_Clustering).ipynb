Case Study : Unsupervised Machine Learning or Clustering.

Old Faithful Geyser, Yellowstone National Park, USA

1. Yellostone National Park in Wyoming state USA is famous for its hot Water natural Geyser called Old Faithful Geyser.

2. Old Faithful Geyser Contains two columns

    * eruption (float) : Eruption Time in minutes.
    * waiting (int) : Waiting Time to between two eruption.

3. The Old Faithful Geyser is an ideal example of Usupervised Machine Learning or Clustering or Exploratory Analysis.

  

```python

# Unsupervied ML Case Study 1 : Old Failthful Geyser
# Dataset contains 2 variables 1. Eruption 2. Waiting time

# example of unsupervised learning or clustering
# old faithful geyser example

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Import the data
df = pd.read_csv(r'/content/faithful_geyser.csv')

# understanding of data set
print(df.shape)
print(df.info())
print(df.head(10))
print(df.tail(10))
print(df.describe())
print(df.dtypes)

# Standardize the data
X_std = StandardScaler().fit_transform(df)

# Run implementation of kmeans
km = KMeans(n_clusters=2, max_iter=100)

km.fit(X_std) # train the dataset
centroids = km.cluster_centers_ # property to get cetroids
print(centroids)
print(len(centroids))

# Plot the clustered data
plt.subplots(figsize=(8, 8), facecolor='lightcoral')

plt.scatter(X_std[km.labels_ == 0, 0], X_std[km.labels_ == 0, 1],
            c='green', label='cluster 1')

plt.scatter(X_std[km.labels_ == 1, 0], X_std[km.labels_ == 1, 1],
            c='blue', label='cluster 2')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=100,
            c='red', label='centroid')

plt.legend()
plt.xlim([-2, 2])
plt.ylim([-2, 2])

print()

plt.xlabel('Eruption time (minutes)',  fontsize=16)
plt.ylabel('Waiting time to next eruption (minutes)', fontsize=16) 
plt.title('Clustered Data from Old Faithful Geyser, \n Yellowstone, USA', \
          fontsize=20, fontweight='bold')
```
